```python
import requests
from bs4 import BeautifulSoup as bs

# Base URL for the Amazon search
base_url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1'
search_query = 'bags'

# Initialize an empty list to store product data
product_data = []

# Loop through multiple pages (20 pages)
for page in range(1, 21):
    params = {
        'k': search_query,
        'page': page
    }

    req = requests.get(base_url, params=params)
    soup = bs(req.text, 'html.parser')

    # Find all product containers on the page
    product_containers = soup.find_all('div', class_='s-result-item')

    for container in product_containers:
        product_name_element = container.find('span', class_='a-text-normal')
        if product_name_element:
            product_name = product_name_element.text
        else:
            product_name = "Not available"

        product_url_element = container.find('a', class_='a-link-normal')
        if product_url_element:
            product_url = 'https://www.amazon.in' + product_url_element['href']
        else:
            product_url = "Not available"

        product_price_element = container.find('span', class_='a-price-whole')
        if product_price_element:
            product_price = product_price_element.text
        else:
            product_price = "Not available"

        rating_element = container.find('span', class_='a-icon-alt')
        if rating_element:
            rating = rating_element.text
        else:
            rating = "Not available"

        num_reviews_element = container.find('span', class_='a-size-base', attrs={"aria-label": " customer reviews"})
        if num_reviews_element:
            num_reviews = num_reviews_element.text
        else:
            num_reviews = "Not available"

        product_data.append({
            'Product URL': product_url,
            'Product Name': product_name,
            'Product Price': product_price,
            'Rating': rating,
            'Number of Reviews': num_reviews
        })

# Print the scraped product data
for i, product in enumerate(product_data, start=1):
    print(f"Product {i}:")
    print("Product URL:", product['Product URL'])
    print("Product Name:", product['Product Name'])
    print("Product Price:", product['Product Price'])
    print("Rating:", product['Rating'])
    print("Number of Reviews:", product['Number of Reviews'])
    print()
```
